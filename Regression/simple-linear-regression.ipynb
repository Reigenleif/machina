```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Salary_Data.csv')
x = df.iloc[:,:1].values
y = df.iloc[:,1:2].values

# Taking care of missing data
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(x[:,:])
x[:,:] = imputer.transform(x[:,:]) 
x = np.array(x, dtype=np.float64)

# Splitting the dataset into the Training set and Test set
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=1,shuffle=True)
```

```python
# Training the Simple Linear Regression model on the Training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()

# Fitting : membuat model dari data training
# fitting dilakukan dengan method regressor.fit()
# regressor.fit(x_train,y_train) menerima 2 parameter, x_train (independent vars) dan y_train (dependent vars)
regressor.fit(x_train,y_train)
```

```python
# Predicting the Test set results

# predict dilakukan dengan method regressor.predict()
# y_pred = regressor.predict(x_test) menerima 1 parameter, x_test (independent vars)
y_pred = regressor.predict(x_test)
```

```python
# Visualising the Training set results
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, regressor.predict(x_train), color = 'blue')
plt.title('(Training set) Salary vs Experience')
plt.ylabel('Salary')
plt.xlabel('Experience')
plt.show()
```

```python
# Visualising the test set results
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_train, regressor.predict(x_train), color = 'blue')
plt.title('(Test Set) Salary vs Experience')
plt.ylabel('Salary')
plt.xlabel('Experience')
plt.show()
```

```python
# getting the value of the equation coefficients
print(regressor.coef_)
print(regressor.intercept_)
```
